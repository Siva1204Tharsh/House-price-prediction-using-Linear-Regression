```python
dataset=pd.read_csv("dataset.csv")
dataset.head(10)
```

```python
dataset.plot(kind='scatter', x='area', y='price')
plt.xlabel("Area")
plt.ylabel("Price of Home")
plt.show()
```

```python
X=dataset.iloc[:,0].values
y=dataset.iloc[:,-1].values
dataset.isnull().any()
```
